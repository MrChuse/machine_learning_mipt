```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import re
%matplotlib inline
```

```python
with open('sentences.txt', 'r') as f:
    l = [l[:-1].lower() for l in f]
l
```

```python
l1 = list(map(lambda t: re.split(r'[^a-z]', t), l))
l2 = [[el for el in s if el] for s in l1]
l2
```

```python
words = set(sum(l2, start=[]))
words_index = {word : i for i, word in enumerate(words)}
words_index # didn't even use it??
```

```python
freq = np.zeros((len(l), len(words)))
for i, sentence in enumerate(l2):
    for j, word in enumerate(words):
        freq[i, j] = sentence.count(word)
freq[0]
```

```python
dists = [(i, scipy.spatial.distance.cosine(freq[0], freqi), l[i]) for i, freqi in enumerate(freq)]
dists.sort(key=lambda x:x[1])
dists
```

```python
with open('out.txt', 'w') as f:
    f.write(str(dists[1][0]) + ' ' + str(dists[2][0]))
```
