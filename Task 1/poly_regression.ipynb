```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return ( np.log(5 * x) - np.cos(x/2) ) / x
#np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
x = np.linspace(1, 15, 100)
plt.plot(x, f(x))
```

```python
def approx(n, points, values):
    assert n == points.shape[0] == points.shape[0]
    left_side = np.array([points ** i for i in range(n)]).T
    weights = np.linalg.solve(left_side, values)
    return weights
```

```python
def construct_poly(weights):
    def poly(x):
        return sum([weight * (x ** i) for i, weight in enumerate(weights)])
    return poly
```

```python
def plot_approx(points):
    w = approx(points.shape[0], points, f(points))
    poly = construct_poly(w)
    plt.plot(x, f(x))
    plt.plot(x, poly(x))
    plt.show()
    return w
```

```python
points = np.array([1, 15])
plot_approx(points)
```

```python
points = np.array([1, 8, 15])
plot_approx(points)
```

```python
points = np.array([1, 4, 10, 15])
w = plot_approx(points)
```
